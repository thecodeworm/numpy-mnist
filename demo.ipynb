# Demoing for the CY Student Showcase

```python
import numpy as np
import matplotlib.pyplot as plt
from network import MultiLayerPerceptron as MLP
from data import X_test, Y_test
import ipywidgets as widgets
from IPython.display import display
```

```python
model = MLP()
model.layer1.W = np.load('weights/layer1W.npy')
model.layer1.b = np.load('weights/layer1b.npy')
model.layer2.W = np.load('weights/layer2W.npy')
model.layer2.b = np.load('weights/layer2b.npy')
```

```python
out = widgets.Output()

def predict_random(b):
    out.clear_output(wait=True)
    with out:
        idx = np.random.randint(0, len(X_test))
        probs = model.forward(X_test[idx:idx+1])
        pred = np.argmax(probs)
        confidence = np.max(probs) * 100
        
        plt.figure(figsize=(4, 4))
        plt.imshow(X_test[idx].reshape(28, 28), cmap='gray')
        plt.title(f"Predicted: {pred}    Confidence: {confidence:.1f}%")
        plt.axis('off')
        plt.show()

button = widgets.Button(description="Random Image")
button.on_click(predict_random)
display(button, out)
```
